```python
#Analyze the date column, and removing entries that have an incorrect timeline

import pandas as pd

df2 = pd.read_csv('311_Service_Requests_from_2010_to_Present.csv', index_col=0, low_memory=False)

# Convert the 'Date' column to datetime with the correct format
df2['Created Date'] = pd.to_datetime(df2['Created Date'], format='%m/%d/%Y %I:%M:%S %p', errors='coerce')
df2['Closed Date'] = pd.to_datetime(df2['Closed Date'], format='%m/%d/%Y %I:%M:%S %p', errors='coerce')


start_date = pd.to_datetime('2000-01-01')
end_date = pd.to_datetime('2023-07-29')

df_filtered = df2[(df2['Created Date'] >= start_date) & (df2['Created Date'] <= end_date) & 
                  (df2['Closed Date'] >= start_date) & (df2['Closed Date'] <= end_date)]

df_filtered
```

```python
#Calculate the time elapsed in closed and creation date

# Convert the 'Created Date' and 'Closed Date' columns to datetime with the correct format
date_format = '%m/%d/%Y %I:%M:%S %p'
df_filtered['Created Date'] = pd.to_datetime(df_filtered['Created Date'], format=date_format, errors='coerce')
df_filtered['Closed Date'] = pd.to_datetime(df_filtered['Closed Date'], format=date_format, errors='coerce')

# Calculate the time difference between 'Closed Date' and 'Created Date'
df_filtered['Time Elapsed'] = df_filtered['Closed Date'] - df_filtered['Created Date']

# Now df_filtered contains a new column 'Time Elapsed' with the calculated time differences

df_filtered
```

```python
df_filtered['Days Elapsed'] = df_filtered['Time Elapsed'].dt.days
df_filtered
```

```python
#4.2 Sort the complaint types based on the average Request_Closing_Time grouping them for different locations

df_filtered_1['Request_Closing_Time'] = df_filtered_1['Time Elapsed']

# Group by location and complaint type
grouped_data = df_filtered_1.groupby(['Location Type', 'Complaint Type'])

# Calculate average Request Closing Time for each group
average_closing_time = grouped_data['Request_Closing_Time'].mean().reset_index()

# Sort complaint types based on average Request Closing Time
sorted_complaints = average_closing_time.sort_values(by='Request_Closing_Time')

# Print sorted complaint types for each location
for location in sorted_complaints['Location Type'].unique():
    location_data = sorted_complaints[sorted_complaints['Location Type'] == location]
    print(f"Location Type: {location}")
    print(location_data[['Complaint Type', 'Request_Closing_Time']])
    print("\n")
```

```python
#6. Identify the significant variables by performing statistical
#analysis using p-values
```

```python
import scipy.stats as stats
```

```python
import pandas as pd
import scipy.stats as stats

# Perform a t-test for each pair of complaint types
complaint_types = avg_response_time.index
significant_vars = []

for i in range(len(complaint_types)):
    for j in range(i+1, len(complaint_types)):
        group1 = df_filtered_1[df_filtered_1['Complaint Type'] == complaint_types[i]]['Request_Closing_Time'].dt.total_seconds()
        group2 = df_filtered_1[df_filtered_1['Complaint Type'] == complaint_types[j]]['Request_Closing_Time'].dt.total_seconds()
        t_stat, p_value = stats.ttest_ind(group1, group2, equal_var=False)  # Assuming unequal variances
        
        if p_value < 0.05:  # Adjust significance level as needed
            significant_vars.append((complaint_types[i], complaint_types[j], p_value))

# Print significant variables
for var in significant_vars:
    print(f"Significant difference between {var[0]} and {var[1]} (p-value: {var[2]})")
```

```python
#7. Perform a Kruskal-Wallis H test
#7.1 Fail to reject H0: All sample distributions are equal
#7.2 Reject H0: One or more sample distributions are not equal

complaint_groups = []
for complaint_type, group in df_filtered_1.groupby('Complaint Type'):
    complaint_groups.append(group['Request_Closing_Time'].dt.total_seconds())

# Perform Kruskal-Wallis H test
h_statistic, p_value = stats.kruskal(*complaint_groups)

# Define significance level
alpha = 0.05

# Compare p-value to significance level
if p_value < alpha:
    print("Reject H0: One or more sample distributions are not equal")
else:
    print("Fail to reject H0: All sample distributions are equal")
    
print(f"Kruskal-Wallis H statistic: {h_statistic}")
print(f"P-value: {p_value}")
```
